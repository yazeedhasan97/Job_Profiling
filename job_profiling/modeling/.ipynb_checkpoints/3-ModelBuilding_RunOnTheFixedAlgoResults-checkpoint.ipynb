```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from joblib import dump, load
import matplotlib.pyplot as plt
import re
import numpy as np
```

```python
# !cd PycharmProjects
df = pd.read_csv('./data/fixed_tree_results/prefinals.csv')
df.info()
```

```python
print(df.isna().sum())
```

```python
total_columns_onehot = 0
total_columns_label = 0
for col in df.columns:
    unq = df[col].unique()
    total_columns_onehot += len(unq)
    total_columns_label += 1
    print(f'The column {col} has {len(unq)} unique values seperated as: {unq}')
    df[col].value_counts().plot(kind='bar')
    plt.show()
print(f'total_columns_onehot = {total_columns_onehot}')
print(f'total_columns_label = {total_columns_label}')
```

```python
onehot_encoded_columns = ['governorate']
encoders = dict()
for col in onehot_encoded_columns:
    encoder = OneHotEncoder()
    temp = pd.DataFrame(
        encoder.fit_transform(df[[col]]).toarray(),
        columns=[name.replace('governorate_', '' ) for name in encoder.get_feature_names_out()]
    )
    df = pd.concat([df, temp], axis=1).drop(col, axis=1)
    encoders[col] = encoder
print(encoders)
df.head()
```

```python
label_encoded_columns = ['education', 'gender']
for col in label_encoded_columns:
    encoder = LabelEncoder()
    df[col] = encoder.fit_transform(df[col])
    encoders[col] = encoder
print(encoders)
df.head()
```

```python
y = df.clusters
X = df.drop(['clusters', 'nodes'], axis=1)
print(X.shape, y.shape)


X_train,X_test, y_train, y_test = train_test_split(X,y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)


for i in range(2, len(X.columns)):
    print("-"*80, i, '-'*80)
    model = DecisionTreeClassifier(max_depth=i)
    model.fit(X_train, y_train)

    train_predictions = model.predict(X_train)

    test_preictions = model.predict(X_test)
    print('Training Confusion Matrix', classification_report(y_train, train_predictions))

    print('Testing Confusion Matrix', classification_report(y_test, test_preictions))
    
```

```python
# I found 6 is the best max_depth
y = df.clusters
X = df.drop(['clusters', 'nodes'], axis=1)
print(X.shape, y.shape)


X_train,X_test, y_train, y_test = train_test_split(X,y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

model = DecisionTreeClassifier(max_depth=6)
model.fit(X_train, y_train)

train_predictions = model.predict(X_train)

test_preictions = model.predict(X_test)
print('Training Confusion Matrix', classification_report(y_train, train_predictions))

print('Testing Confusion Matrix', classification_report(y_test, test_preictions))
```

```python
for encoder in encoders:
    dump(encoders[encoder], f'./runs/{encoder}_encoder.joblib')
dump(model, './runs/model.joblib')
```
